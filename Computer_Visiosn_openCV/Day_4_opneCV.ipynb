```python
# 14:  how to draw lines, and shapes in python

import cv2 as cv
import numpy as np

# draw a canvas

imgb =np.zeros((600,600)) #black for zeros
imgw = np.ones((600,600)) # white for ones

# print size
print("The size of our Canvas is: ", imgb.shape)
# print(imgb)

# adding colors to canvas
colored_img = np.zeros((600,600,3), np.uint8) # color channel addition
colored_img[:] = 255,0,255 #purple color key #complete color img
colored_img[150:230, 100:500] = 29,00,800 # part of image to be colored

# adding line

cv.line(colored_img, (0,0),(300,300), (255,0,0),3) #partline
cv.line(colored_img,(0,0),(colored_img.shape[0],colored_img.shape[1]), (255,0,0),3) #crossed line

#drawing rectangle
cv.rectangle(colored_img,(50,100),(300,400),(244,65,89),8) # drawing rectangle
cv.rectangle(colored_img,(50,100),(300,400),(244,65,89),cv.FILLED) # filled rectangle

# drawing circle
cv.circle(colored_img, (400,300), 50, (0,78,98), 6) # circle with edges
cv.circle(colored_img, (200,300), 50, (0,78,98), cv.FILLED) # circle with filled

# Adding text

cv.putText(colored_img, "How Are You?", (200,400), cv.FONT_HERSHEY_DUPLEX, 1,(2,255,0),3)

# cv.imshow("Black",imgb)
# cv.imshow("White",imgw)
cv.imshow("Colored", colored_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Saving HD REcording of cam stream

# 15: managing Resolution of cam

import cv2 as cv
import numpy as np

cap = cv.VideoCapture(0)

def hd_resolution():
    cap.set(3,1280)
    cap.set(4,720)

def sd_resolution():
    cap.set(3,640)
    cap.set(4,480)

def fhd_resolution():
    cap.set(3,1920)
    cap.set(4, 1080)

def fps30():
    cap.set(cv.CAP_PROP_FPS, 5)

    
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out= cv.VideoWriter("Resources/Rec_video.avi", cv.VideoWriter_fourcc("M","J","P", "G"),10,(frame_width,frame_height))

while(True):
    (ret, frame) = cap.read()
    # to show in player
    if ret == True:
        out.write(frame)
        cv.imshow(frame)
        #to quit with q key
        if cv.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break
out.release()
cap.release()
cv.destroyAllWindows()
```
