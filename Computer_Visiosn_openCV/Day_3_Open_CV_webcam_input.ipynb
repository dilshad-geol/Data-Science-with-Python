```python
# 09: how to capture a webcam inside python

# step 1 Import libraries
import cv2 as cv
import numpy as np
# Step 2 :Read the frame from camera 
cap = cv.VideoCapture(0) #webcam no. 1, webcame no. 2 if (1)
# if (cap.isOpened() == False):
    # print("error loading webcam")
# read untill the end
# Step 3 Display frame by frame
while(cap.isOpened()):
    #capture frame by frame
    ret, frame =cap.read()
    if ret == True:
        # to diplay frames
        cv.imshow("frame", frame)
        if cv.waitKey(1) & 0xFF ==ord("q"):
            break
    else:
        break
# realease or close windows easily
cap.release()
cv.destroyAllWindows()
```

```python
# 10: convert camera colors in webcam inside python


import cv2 as cv
import numpy as np
cap = cv.VideoCapture(0)

while(True):
    (ret, frame) = cap.read()
    gray_frame= cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    (thresh, binary) =cv.threshold(gray_frame, 127, 255, cv.THRESH_BINARY)
    cv.imshow("Original Cam", frame)
    cv.imshow("Gray Cam", gray_frame)
    cv.imshow("Binary Cam", binary)
    if cv.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break
cap.release()
cv.destroyAllWindows()
```

```python
# 11: Writing videos from cam

import cv2 as cv
import numpy as np
cap = cv.VideoCapture(0)

# writing format,codec,video writer object, file output

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out= cv.VideoWriter("Resources/Rec_video.avi", cv.VideoWriter_fourcc("M","J","P", "G"),10,(frame_width,frame_height))

while(True):
    (ret, frame) = cap.read()
    grayframe= cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    (thresh, binary) =cv.threshold(gray_frame, 127, 255, cv.THRESH_BINARY)
    # to show in player
    if ret == True:
        out.write(grayframe)
        cv.imshow("Original Cam", frame)
        cv.imshow("Gray Cam", grayframe)
        cv.imshow("Binary Cam", binary)
        #to quit with q key
        if cv.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break
    out.release
cap.release()
cv.destroyAllWindows()
```

```python
# 12: Setting of camera or video (brightness, height, width)

import cv2 as cv
import numpy as np

cap= cv.VideoCapture(0) #capture webcam

cap.set(10,50) # 10 is the key to set brightness
cap.set(3, 640) # width key(3), set value to 640 (changeable)
cap.set(4,480) # hieght key (4), set value to 480 (changeable)

while (True):
    ret,frame = cap.read()
    if ret ==True:
        cv.imshow("frame", frame)
        if cv.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break    
cap.release()
cv.destroyAllWindows()
```

```python
# 13: Basic functions or manipulations in openCV

import cv2 as cv
import numpy as np
img =cv.imread("Resources/img1.jpg")

# resize
resize_img = cv.resize(img, (650, 250))#(width,height)

#gray
gray_img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

# blurred img
blur_img = cv.GaussianBlur(img, (9,9), 0) #intensity control (7,7) it should always be odd

# edge detection
edge_img = cv.Canny(img, 48,48) #canny edge detection

# thickness of lines
mat_kernel =np.ones((7,7),np.uint8) # control edge thickness
dilated_img =cv.dilate(edge_img, (mat_kernel), iterations=1)

# make thinner outlines (erode)
ero_img= cv.erode(dilated_img, mat_kernel, iterations=1)

# Black_n_white binary 
(thresh, b_w)= cv.threshold(gray_img,127,255, cv.THRESH_BINARY)

# Cropping we will use numpy not openCV
crop_img = img[0:200, 200:300]

cv.imshow("Original", img)
cv.imshow("Resized", resize_img)
cv.imshow("Gray", gray_img)
cv.imshow("Blurred", blur_img)
cv.imshow("Edge", edge_img)
cv.imshow("Dilated Lines", dilated_img)
cv.imshow("Eroded Img", ero_img)
cv.imshow("Binary",b_w)
cv.imshow("Cropped", crop_img)

    

cv.waitKey(0)
cv.destroyAllWindows()
```
