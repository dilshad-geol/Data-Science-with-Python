```python

#Box_PLot
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from numpy import mean
#import data
df = sns.load_dataset("titanic")
df
#Setting figure size
plt.figure(figsize=(10,10))
#setting style
sns.set_style(style = None, rc= None)
sns.set_style("white")
#setting pellet
#setting color
#plotting
sns.barplot(x="class", y="fare", data=df,
            linewidth =2.5, facecolor=(0,0,1.0,0.1),
            errcolor=".2", edgecolor="0.2")

#title
plt.title("Titanic")
#Labelling
plt.xlabel("Sex")
plt.ylabel("Travelling Alone")
#plot show
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
kashti= sns.load_dataset("titanic")
kashti.head()
sns.boxplot(x="survived",y="age",hue="sex", dodge= True, data= kashti, showmeans=True, meanprops = {"marker":"*", "markersize":"12","markeredgecolor":"red"})
plt.title("Kitny bach gye, kitny Doob gye", size=14)
plt.xlabel("Doob Gye", size=10, weight="bold")
plt.ylabel("Age",size = 10, weight= "bold")
plt.show()
```

# Lineplot with multifacets

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df= sns.load_dataset("dots")
df
p= sns.color_palette("rocket_r")
sns.relplot(x="time",y="firing_rate",data=df ,hue="coherence", col="align", kind="line",size_order=["T1","T2"],
           palette = p,height= 5,aspect=.75,facet_kws=dict(sharex=False) )
plt.show
```
