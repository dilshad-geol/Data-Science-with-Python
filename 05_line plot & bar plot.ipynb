```python
#import llibraries
import seaborn as sns
import matplotlib.pyplot as plt
#import data
phool =sns.load_dataset("iris")
phool
#figure size
plt.figure(figsize=(20,16))
sns.set_style(style = None, rc= None)
sns.set_style("whitegrid")
#plotting data
sns.lineplot(x="sepal_length", y="sepal_width", data= phool)
#adding title
plt.title("Phoolon ka lineplot")
#adding limit
plt.xlim(6.0)
plt.ylim(2.6)
#setting axis labels
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
#show plot
plt.show()
```

```python
#impport libraries
import seaborn as sns
import matplotlib.pyplot as plt
#import data
brpltdata= sns.load_dataset("iris")
brpltdata
#Seeting Style
sns.set_style(style = None , rc= None)
sns.set_style("dark")
#plotting data
sns.barplot(x="species", y="sepal_width", data = brpltdata)
#Setting title
plt.title("Phoolon ka Barplot")
#changing labels
plt.ylabel("Sepal Width")
plt.xlabel("Species")
#plot show
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#import data
df = sns.load_dataset("titanic")
df
#Setting figure size
plt.figure(figsize=(10,10))
#setting style
sns.set_style(style = None, rc= None)
sns.set_style("white")
#plotting
sns.barplot(x="sex", y="alone", hue= "who", data=df, order = ("female", "male"), color = (None))
#title
plt.title("Titanic")
#Labelling
plt.xlabel("Sex")
plt.ylabel("Travelling Alone")
#plot show
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from numpy import mean
#import data
df = sns.load_dataset("titanic")
df
#Setting figure size
plt.figure(figsize=(10,10))
#setting style
sns.set_style(style = None, rc= None)
sns.set_style("white")
#setting pellet
#setting color
#plotting
sns.barplot(x="sex", y="alone", hue= "who", data=df, order = ("female", "male"), ci=None, palette = "flare", estimator= mean, saturation = 0.4)
#title
plt.title("Titanic")
#Labelling
plt.xlabel("Sex")
plt.ylabel("Travelling Alone")
#plot show
plt.show()
```
