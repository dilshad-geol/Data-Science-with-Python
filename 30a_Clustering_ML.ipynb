# Clustering (Unsupervised Machine Learning)

Clustering: Grouping of individuals based on their common characteristics
- Segmentation.....CLustering(Same)
  - Each cluster have something in common characteristics
  - Create profile of each group
    - e.g. bank data clustering into age, money,salary etc.
  - based on the simimlarities to eachother,
    - grouping
  - 'Customer segmentation' is one of popular usage of clustering.

Def: A group of object that are similar to other subjects in the cluster and dissimilar to data points in other clusters. 

 _Clustering_ vs _Classification_
|Clustering|Classification|
|---|---|
|unsupervised|supervised|
|No training data|training data|
|unlabelled data|labeled data(each data lable belongs to a particular class)|
|define label using data algorithms(k-means)|data--> training--> classification|
|Segmentation||

- Summary generation

**Clustering Algorithms**
- Partitioned based clustering
  - K-means
  - fuzzy c-means
- Hierarchical based clustering
  - trees of clusters
    - Agglomerative
    - Divisive
- Density based clustering
  - produces arbitary shaped clusters
    - good for dealing with spatial clusters
    - or when there is noise in dataset
      - DBSCAN
