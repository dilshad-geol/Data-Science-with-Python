## Partitioning Methods
- We split the data into smaller subsets.
  - **Centroid based(e.g. K-means)**
    - the centre of the cluster could not be a part of the data
  - **Medoids-based(k-medoids)**
    - the centre of the cluster must be a point in the dataset. 


- **K-Means Algorithm**
  - Randomly pick k centroids
  - Add each instance to the closest centroid
  - Recompute the centroids(the centroid is the centre i.e., "average" point of the cluster)
  - Add each instance to the closest centroid.
    - Recompute the centroids(average)
    - Repeat until no changes occur anymore.

- **K-means**
  - sensitive to Outliers
  - Converges at local minima
  - Number of clusters to be known

- **K-medioid**
  - the centre of the cluster must be a point in the dataset.
  - Robust against Outliers
## Hierarchical Methods
- **Agglomerative Clustering**
  - No need to specify the number of clusters
- **Density based Clustering**
  - DB Scan
    - Can find clusters in any shape
    - Used to cluster high dimensional data
- **EM Clustering**
  - expectation maximum
