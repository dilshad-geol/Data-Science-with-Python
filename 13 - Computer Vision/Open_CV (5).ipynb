```python
# Assignmnet: how to define a function to stack multiple images of different sizes and tones

import cv2 as cv
import numpy as np


img = cv.imread("Resources/img1.jpg")
img1 = cv.imread("Resources/img2.jpg")
img2 = cv.imread("Resources/img3.jpg")
# Stacking same image

#1- Horizontal Stack:

hor_img= np.hstack((img,img))

#2- Vertical Stack

ver_img = np.vstack((img,img))


# define a function for vertically concatenating images of different widths 
def vconcat_resize(img_list, interpolation 
                   = cv.INTER_CUBIC):
      # take minimum width
    w_min = min(img.shape[1] 
                for img in img_list)
      
    # resizing images
    im_list_resize = [cv.resize(img,
                      (w_min, int(img.shape[0] * w_min / img.shape[1])),
                                 interpolation = interpolation)
                      for img in img_list]
    # return final image
    return cv.vconcat(im_list_resize)
# function calling
img_v_resize = vconcat_resize([img, img2, img1])
        

cv.imshow("Vertical_stack_resized", img_v_resize)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# define a function for horizontally concatenating images of different heights 

import cv2 as cv
import numpy as np
img = cv.imread("Resources/img1.jpg")
img1 = cv.imread("Resources/img2.jpg")
img2 = cv.imread("Resources/img3.jpg")

# defining Function
def hconcat_resize(img_list, 
                   interpolation 
                   = cv.INTER_CUBIC):
      # take minimum hights
    h_min = min(img.shape[0] 
                for img in img_list)
      
    # image resizing 
    im_list_resize = [cv.resize(img,
                       (int(img.shape[1] * h_min / img.shape[0]),
                        h_min), interpolation
                                 = interpolation) 
                      for img in img_list]
      
    # return final image
    return cv.hconcat(im_list_resize)
  
# function calling
img_h_resize = hconcat_resize([img, img2, img1])
  
# show the Output image
cv.imshow('hconcat_resize.jpg', img_h_resize)

cv.waitKey(0)
cv.destroyAllWindows()
```
