|----|----|

**Variable Type -1**
- **Categorical(Nominal)**:
    - Binominal: simply 2 values or relationship discussed(true or false, yes or no)
    - Multinominal: different number of values(Multiple options in travelling)

**Variable Type -2**
- **Categorical(Nominal)**:
    - Ordinal variable: Data  ranked or ordered, how much shirts i have?
        - Categories can be compared and no fix unit of measurement for statistics

**Variable Type -3**
- **Ratio Data**:
    - data have a natural zero.
    - Measurement in units and ratios are **Continous Variable**
        - e.g. today`s sale was good than yesterday (economic data)

**Variable Type -4**
- **Interval/ Variable Data**:
    - Ordered and characterized data.
    - Ratios are meaning-less (50 degrees is not double hot of 25 degrees)
    - Differences are meaningful.

###  Notions and Terms

**N = size of population**
**n= size of sample**
**Sigma = Sum**

- A **measurement** is a number or attribute computed for each member of a population or of a sample. The measurements of sample elements are collectively called sample data.

- A **parameter** is a number that summarizes some aspect of the population as a whole.
- A **statistic** is a number computed from the sample data.
- **Descriptive statistics** is the branch of statistics that involves organizing, diplaying, and describing data.
- **Inferential statistics** is the branch of statistics that involves drawing conclusions about a population based on information contained in a sample taken from the population.
- A **population** is any specific collection of objects of interest.
- Any **sample** is any subset or sub-collection of population, including the case that the sample connsists of the whole population, in which case it is termed a census.
- **Qualitative data** are measurements for which there is no natural numerical scale, but which consists of attributes, labels, or other non-numerical characteristics.
- **Quantitative data** are numerical measurements that arise from a natural numerical scale.
- **Mean** is the sum divided by the number of observations(Average). The mean is the average or the most common value in a collection of numbers.
      - **meaningful for intervals and ratio data**
      - **Outliers** change the means of a data, therefore **median** is useful.
      - **Population mean** = mu = sigma x/ N
      - **Sample mean** = x bar = sigma x/ n
- **Median** is the middle number in a sorted, ascending or descending, list of numbers.
      - The median is unique for each dataset
      - Outliers dont have any effect on median
      - Ratios, intervals and ordinal data has best use.
- **Mode** The value that occurs most frequently e.g. most common age in a group of people in a class.

##  Measure of Dispersion
- **variability**
- **scatter or spread**

**Dispersion**: How much data is spread around mean? kniwn as;
                    -   standard deviation(std)
                    -   standard error(se)
                    -   variance
                    -   Bell curve
- Caused by **differernce** between the **minimum** and the **maximum**
    - **range** = **minimum**-to-**maximum**
                    -   standard deviation(std)
                    -   standard error(se)
                    -   variance
                    -   Bell curve

**Example:**
prices= (30,30, 35,45,10,61,70,90,115)

Mean = 54

Median = 45

Mode = 30

std= 33.26

variance = 1106.5 (std= underroot of variance)

*std has same units as sample and population.*

min = 45

max= 115

**Standard Deviation(SD) & Mean**: 
- **Mean** only gives us a small picture
- **Means** are incomplete without **dispersion(SD)**
- **Mean** with a **Standard Deviation** is more useful than only mean by it self.
