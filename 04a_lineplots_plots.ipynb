## Adding limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load data sets
data1 = sns.load_dataset("car_crashes")
data1
#draw a lineplot
sns.lineplot(x="ins_premium", y ="alcohol", data=data1)
#Adding a title
plt.title("Line Plot")
#x-Limit
plt.xlim(700)
# #y limit
plt.ylim(4)

#plotting
plt.show()
```

# Adding styles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load data sets
data1 = sns.load_dataset("iris")
data1
#setting Figure Size
plt.figure(figsize=(10,8))
#setting style
sns.set_style(style = None, rc=None)
sns.set_style('dark')
#draw a lineplot 
sns.lineplot(x="sepal_length", y ="sepal_width",  data=data1)
#Adding a title
plt.title("Line Plot")
# Labelling x-axis and y-axis
plt.ylabel("Sepal Width")
plt.xlabel("Sepal Length")
#addung style
#plotting
plt.show()
```
